```python
#importing tools
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
#importing dataframes
analysis_df = pd.read_excel("analysis_df.xlsx")

analysis_df.head()
```

## Plotting Graphs

Spain's monthly growth in nominal exchange rates versus US Dollar

```python
#Plotting nominal exchange rates
plt.plot(analysis_df["Date"], analysis_df["Spain Monthly Difference in Log Exchange Rate"])

#Plotting date spain left Bretton Woods
plt.axvline(x = analysis_df.loc[139, "Date"], label = "Date of Leaving Bretton Woods (1971-08-01)", linestyle = 'dotted', color = 'r')

#Labelling and applying legend
plt.title("Spain's Monthly Growth in Nominal Exchange Rates Versus The US Dollar")
plt.xlabel("Date")
plt.ylabel("Percentage Change to Monthly Exchange Rate")
plt.legend()

#Adjusting plot size for clarity of view
plt.rcParams["figure.figsize"] = (10,5)


plt.show()
```

Spain's Monthly growth in real exchange rates versus the US Dollar

```python
#Plotting real exchange rates
plt.plot(analysis_df["Date"], analysis_df["Real Exchange Rate Growth"])

#Plotting date spain left Bretton Woods
plt.axvline(x = analysis_df.loc[139, "Date"], label = "Date of Leaving Bretton Woods (1971-08-01)", linestyle = 'dotted', color = 'r')

#Labelling and applying legend
plt.title("Spain's Monthly Growth in Real Exchange Rates Versus The US Dollar")
plt.xlabel("Date")
plt.ylabel("Percentage Change to Monthly Exchange Rate")
plt.legend()

plt.show()
```

Plotting Real and Nominal rate changes against eachother

```python
#Plotting nominal exchange rates
plt.plot(analysis_df["Date"], analysis_df["Spain Monthly Difference in Log Exchange Rate"], label = 'Nominal Rate Changes')

#Plotting real exchange rates
plt.plot(analysis_df["Date"], analysis_df["Real Exchange Rate Growth"], label = 'Real Rate Changes')

#Plotting date spain left Bretton Woods
plt.axvline(x = analysis_df.loc[139, "Date"], label = "Exit of Bretton Woods (1971-08-01)", linestyle = 'dotted', color = 'r')

#Labelling and applying legend
plt.title("Spain's Monthly Growth in Exchange Rates Versus The US Dollar")
plt.xlabel("Date")
plt.ylabel("Percentage Change to Monthly Exchange Rate")
plt.legend(loc = 'lower right')

plt.show()
```

Plotting Monthly indexed values of international gold reserves

```python
#Plotting spanish index, until exit
plt.plot(analysis_df.loc[:139,"Date"], analysis_df.loc[:139, 'Spain Index of International Reserves'])

#Labelling
plt.title("Spanish Index of International Gold Reserves Under Bretton Woods")
plt.xlabel("Date")
plt.ylabel("Index")

plt.show()
```

```python
#Plotting American Index, until exit
plt.plot(analysis_df.loc[:139, "Date"], analysis_df.loc[:139, 'US Index of International Reserves'])

#Labelling
plt.title("US Index of International Gold Reserves Under Bretton Woods")
plt.xlabel("Date")
plt.ylabel("Index")

plt.show()
```
