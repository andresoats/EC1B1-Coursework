```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import datetime
```

```python
analysis_df = pd.read_excel("analysis_df.xlsx")

analysis_df.head()
```

# Spain's Monthly Rate of Inflation

```python
#Plotting inflation rates
plt.plot(analysis_df["Date"], analysis_df["Spain Monthly Inflation Rate %"])

#Plotting date Spain left Bretton Woods
plt.axvline(x = analysis_df.loc[139, "Date"], label = "Date of Leaving Bretton Woods (1971-08-01)", linestyle = 'dotted', color = 'r')

#Labels
plt.title("Spain's Monthly Rate of Inflation")
plt.xlabel("Date")
plt.ylabel("Rate of Inflation (%)")
plt.legend()

plt.show()
```

# Spain's Monthly Difference in Log Industrial Production 

```python
#Plotting log of industrial production
plt.plot(analysis_df["Date"], analysis_df["Spain Monthly Difference in Log Industrial Production"])

#Plotting date Spain left Bretton Woods
plt.axvline(x = analysis_df.loc[139, "Date"], label = "Date of Leaving Bretton Woods (1971-08-01)", linestyle = 'dotted', color = 'r')

#Labels
plt.title("Spain's Monthly Difference in Log Industrial Production")
plt.xlabel("Date")
plt.ylabel("Log of Industrial Production")
plt.legend()

plt.show()
```

```python
#Plotting growth in industrial production
plt.plot(analysis_df["Date"], analysis_df["Spain Growth in industrial production versus 12 months ago % change"])

#Plotting date Spain left Bretton Woods
plt.axvline(x = analysis_df.loc[139, "Date"], label = "Date of Leaving Bretton Woods (1971-08-01)", linestyle = 'dotted', color = 'r')

#Labels
plt.title("Spain's Monthly Growth of Industrial Production versus 12 Months Ago")
plt.xlabel("Date")
plt.ylabel("Growth of Industrial Production versus 12 Months Ago (%)")
plt.legend()

plt.show()
```
